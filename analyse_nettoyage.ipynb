CorrÃ©lations avec la Target

```python
df_man_no_smoking = df[(df["is_male"]== 1) & (df["is_smoker"]== 0)]
df_woman_no_smoking = df[(df["is_male"]== 0) & (df["is_smoker"]== 0)]
df_man_smoking = df[(df["is_male"]== 1) & (df["is_smoker"]== 1)]
df_woman_smoking = df[(df["is_male"]== 0) & (df["is_smoker"]== 1)]


data_man_no_smoking = np.array(df_man_no_smoking, dtype=float)
data_woman_no_smoking = np.array(df_woman_no_smoking, dtype=float)
data_man_smoking = np.array(df_man_smoking, dtype=float)
data_woman_smoking = np.array(df_woman_smoking, dtype=float)

plt.xlabel("age")
plt.ylabel("charges")
plt.plot(data_man_no_smoking[:, 0], data_man_no_smoking[:, 3], "c.")
plt.plot(data_woman_no_smoking[:, 0], data_woman_no_smoking[:, 3] ,"y.")
plt.plot(data_man_smoking[:, 0], data_man_smoking[:, 3] ,"b.")
plt.plot(data_woman_smoking[:, 0], data_woman_smoking[:, 3] ,"r.")

plt.show()
```

```python
df_is_obese= df [df["bmi"]>24.9]
df_is_obese_and_smoking = df [(df["bmi"]>24.9) & (df["is_smoker"]==1)]

bmi_obese=df_is_obese["bmi"].values
charges_obese= df_is_obese["charges"].values
bmi_obese_smoker=df_is_obese_and_smoking["bmi"].values
charges_obese_smoker=df_is_obese_and_smoking["charges"].values

plt.xlabel("bmi")
plt.ylabel("charges")
plt.title("Relation entre BMI smoker et charges")
plt.scatter(bmi_obese_smoker, charges_obese_smoker, alpha=0.7, c="purple", label="ObÃ¨ses smoker")  
plt.scatter(bmi_obese, charges_obese, alpha=0.4, c="pink", label="ObÃ¨ses")  
plt.legend()
plt.show()
```
