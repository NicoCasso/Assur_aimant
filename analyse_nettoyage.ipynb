```python
df['is_male'] = df['sex'].str.startswith('m').astype(int)
df['is_female'] = df['sex'].str.startswith('f').astype(int)
df['is_smoker'] = df['smoker'].str.startswith('y').astype(int)
df['is_southwest']= df ['region'].str.startswith('s').astype(int)
df['is_northwest']= df ['region'].str.startswith('n').astype(int)

df
```

```python
df = df.drop("sex",axis=1)
df = df.drop("smoker",axis=1)
df = df.drop("region",axis=1)
df
```

Analyse bivariÃ©e (corrÃ©lation entre les variables)

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='is_smoker', y='age', data=df)
plt.xlabel('Is Smoker (0: Non-fumeur, 1: Fumeur)')
plt.ylabel('Age')
plt.title('Distribution of Age by Smoker Status')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='is_smoker', y='children', data=df)
plt.xlabel('Is Smoker (0: Non-fumeur, 1: Fumeur)')
plt.ylabel('Number of children')
plt.show()

#pas bon ca
```
